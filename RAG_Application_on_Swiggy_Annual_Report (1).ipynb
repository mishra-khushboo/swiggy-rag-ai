```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("swiggy_annual_report.pdf")
documents = loader.load()

print(len(documents))
```

```python
!pip install -U langchain langchain-community langchain-huggingface langchain-text-splitters faiss-cpu pypdf sentence-transformers
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

docs = text_splitter.split_documents(documents)

print("Number of chunks:", len(docs))
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
from langchain_community.vectorstores import FAISS

vectorstore = FAISS.from_documents(docs, embeddings)
```

```python
retriever = vectorstore.as_retriever(search_kwargs={"k":3})
```

```python
query = "What services does Swiggy offer?"

results = retriever.invoke(query)

for doc in results:
    print(doc.page_content)
    print("------")
```

```python
context = " ".join([doc.page_content for doc in results])

print(context)
```

```python
def ask_question(question):

    # Retrieve relevant chunks
    results = retriever.invoke(question)

    # Combine context
    context = " ".join([doc.page_content for doc in results])

    print("Question:", question)
    print("\nRelevant Context:\n")
    print(context)

    print("\nAnswer:\n")
    print(context[:500])   # simple answer from document
```

```python
ask_question("What services does Swiggy offer?")
```

```python
ask_question("Who founded Swiggy?")
```

```python
ask_question("What is Swiggy Instamart?")
```

```python
ask_question("What are Swiggy's business segments?")
```

```python
while True:
    question = input("Ask a question (type exit to stop): ")

    if question.lower() == "exit":
        break

    ask_question(question)
```

```python
from transformers import pipeline

llm = pipeline(
    "text-generation",
    model="google/flan-t5-base",
    max_length=512
)
```

```python
def generate_answer(question, context):

    prompt = f"""
You are an AI assistant answering questions ONLY using the provided context from the Swiggy Annual Report.

If the answer is not present in the context, say:
"I could not find this information in the Swiggy Annual Report."

Context:
{context}

Question:
{question}

Answer:
"""

    result = llm(prompt)[0]["generated_text"]
    return result
```

```python
def ask_question(question):

    docs = retriever.invoke(question)

    context = "\n\n".join([doc.page_content for doc in docs])

    answer = generate_answer(question, context)

    print("\nQuestion:", question)

    print("\nAnswer:")
    print(answer)

    print("\nSource Pages:")
    for doc in docs:
        print(doc.metadata)
```

```python
ask_question("What services does Swiggy offer?")
```

```python
while True:

    query = input("\nAsk a question about Swiggy Annual Report (type exit to stop): ")

    if query.lower() == "exit":
        break

    ask_question(query)
```
